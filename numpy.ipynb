Q2

```python
import numpy as np
arr2 = np.array([1.5, 2.6, 3.7, 4.8])
print("转换前数组:", arr2)
print("转换前数据类型:", arr2.dtype.name)

arr2_int32 = arr2.astype(np.int32)
print("\n转换后数组:", arr2_int32)
print("转换后数据类型:", arr2_int32.dtype.name)
```

Q4

```python
import numpy as np
arr = np.array([10, 15, 20, 25, 30, 35])
r1 = arr[[0, 2, 4]]
print("花式索引结果:", r1)
r2 = arr[arr > 20]
print("布尔索引结果:", r2)
```

Q6

```python
import numpy as np
x = np.array([1, 4, 9, 16])
y = np.array([2, 1, 4, 5])
sqrt_x = np.sqrt(x)
print("x的平方根:", sqrt_x)
max = np.maximum(sqrt_x, y)
print("对应位置最大值:", max)
```

Q8

```python
import numpy as np
arr_x = np.array([1, 5, 7])
arr_y = np.array([2, 6, 8])
arr_con = np.array([True, False, True])
r = np.where(arr_con, arr_x, arr_y)
print("条件逻辑运算结果:", r)
```

Q9

```python
import numpy as np
arr = np.array([[6, 2, 7], [3, 6, 2], [4, 3, 2]])
m = arr.mean()
print("数组所有元素的算术平均值:", m)
arr.sort(axis=0)
print("按列排序后的数组:")
print(arr)
arr.mean(axis=1)
```

```python
import numpy as np
np.random.seed(10)
A = np.random.rand(2, 2)
B = np.random.rand(2, 2)
print("矩阵A:")
print(A)
print("\n矩阵B:")
print(B)
dot_product = np.dot(A, B)
print("\n矩阵A和B的点积结果:")
print(dot_product)
```
