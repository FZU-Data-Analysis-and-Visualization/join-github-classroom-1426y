```python
df_group_score = df1.groupby("gender")["score"].agg(["mean", "max"])
df_group_score.columns = ["分数均值", "分数最大值"]
df_group_score
```

```python
df_group_count = df1.groupby("gender").size().reset_index(name="人数")
df_group_count
```
